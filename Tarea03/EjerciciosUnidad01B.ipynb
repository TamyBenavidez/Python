1. Utilice aritm√©tica de corte de tres d√≠gitos para calcular las siguientes sumas. Para cada parte, ¬øqu√© m√©todo es m√°s preciso y por qu√©?

a. $\sum_{i=1}^{10} \left ( \frac{1}{i^2} \right)$ primero por: $\frac{1}{1}+ \frac{1}{4} + \cdots + \frac{1}{100}$ Y luego por: $\frac{1}{100} + \frac{1}{81} + \cdots + \frac{1}{1}$

```python
import math

# Funci√≥n para cortar a 3 cifras significativas
def truncar_3(x):
    if x == 0:
        return 0.0
    potencia = math.floor(math.log10(abs(x)))
    factor = 10 ** (2 - potencia)
    return math.trunc(x * factor) / factor

# Suma con aritm√©tica de corte a 3 cifras significativas
def suma_truncada(valores, mostrar=False):
    s = 0.0
    for v in valores:
        s = truncar_3(s + v)
        if mostrar:
            print(f"Suma parcial: {s}")
    return s

# Sumatoria de 1/i¬≤
valores_a = [truncar_3(1/i**2) for i in range(1, 11)]

print("Valores truncados (1/i¬≤):")
print(valores_a)

print("\nSuma ascendente (1 a 10)")
suma_asc_a = suma_truncada(valores_a, mostrar=True)
print("Suma ascendente =", suma_asc_a)

print("\nSuma descendente (10 a 1)")
suma_desc_a = suma_truncada(list(reversed(valores_a)), mostrar=True)
print("Suma descendente =", suma_desc_a)

# Comparar con valor real
real_a = sum(1/i**2 for i in range(1, 11))
print(f"\nValor real (doble precisi√≥n): {real_a:.6f}")

print("\nErrores absolutos:")
print(f"Ascendente: {abs(real_a - suma_asc_a)}")
print(f"Descendente: {abs(real_a - suma_desc_a)}")
```

### b.  $\sum_{i=1}^{10} \left( \frac{1}{i^3} \right)$ primero por: $\frac{1}{1} + \frac{1}{8} + \frac{1}{27} + \cdots + \frac{1}{1000}$ Y luego por: $\frac{1}{1000} + \frac{1}{729} + \cdots + \frac{1}{1}$

```python
# Sumatoria de 1/i¬≥ 
valores_b = [truncar_3(1/i**3) for i in range(1, 11)]

print("Valores truncados (1/i¬≥):")
print(valores_b)

print("\nSuma ascendente (1 a 10)")
suma_asc_b = suma_truncada(valores_b, mostrar=True)
print("Suma ascendente =", suma_asc_b)

print("\nSuma descendente (10 a 1)")
suma_desc_b = suma_truncada(list(reversed(valores_b)), mostrar=True)
print("Suma descendente =", suma_desc_b)

# Comparar con valor real
real_b = sum(1/i**3 for i in range(1, 11))
print(f"\nValor real (doble precisi√≥n): {real_b:.6f}")

print("\nErrores absolutos:")
print(f"Ascendente: {abs(real_b - suma_asc_b)}")
print(f"Descendente: {abs(real_b - suma_desc_b)}")
```

Se puede confirmar que la suma ascendente es la m√°s precisa, porque los t√©rminos peque√±os no se pierden por el corte o truncamiento.
La diferencia entre las sumas muestra c√≥mo los errores se van acumulando seg√∫n el orden de las operaciones.

2. La serie de Maclaurin para la funci√≥n arcotangente converge para ‚àí1 < ùë• ‚â§ 1 y est√° dada por

   $arctan x = \lim_{n \to \infty} P_n(x) = lim_{n \to \infty} \sum_{i=1}^{n} (-1)^{i+1} \frac{x^{2i-1}}{2i-1}$

a. Utilice el hecho de que tan ùúã‚ÅÑ4 = 1 para determinar el n√∫mero n de t√©rminos de la serie que se necesita sumar para garantizar que $|4ùëÉùëõ(1) ‚àí ùúã| < 10^{‚àí3}$

```python
x = 1
tol = 1e-3
n = 1

while True:
    Pn = sum(((-1)**(i+1)) * x**(2*i-1)/(2*i-1) for i in range(1,n+1))
    error = abs(4*Pn - 3.141592653589793)  # ùúã real
    if error < tol: break
    n += 1

print("n para |4Pn(1)-ùúã|<1e-3:", n)
print("Aproximaci√≥n de ùúã:", 4*Pn)
```

3. Otra f√≥rmula para calcular ùúã se puede deducir a partir de la identidad $\frac{ùúã}{4} = 4 arctan \frac{1}{5} - arctan \frac{1}{239}  $. Determine el n√∫mero de t√©rminos que se deben sumar para garantizar una aproximaci√≥n ùúã dentro de $10^{-3}$.

```python
import math

x1, x2 = 1/5, 1/239
tol = 1e-3
n = 1

while True:
    error = 16*x1**(2*n+1)/(2*n+1) + 4*x2**(2*n+1)/(2*n+1)
    if error < tol: break
    n += 1

Pn = lambda x: sum(((-1)**(i+1)) * x**(2*i-1)/(2*i-1) for i in range(1,n+1))
pi_approx = 4*(4*Pn(x1) - Pn(x2))
print("T√©rminos necesarios:", n)
print("œÄ aproximado:", pi_approx)
print("Error absoluto:", abs(math.pi - pi_approx))
```

* Se puede observar el valor m√≠nimo de ùëõ
* La f√≥rmula de Machin converge muy r√°pido debido al t√©rmino (1/239)
* El ùëõ por la cota es peque√±o.

1.	Escriba un algoritmo para sumar la serie finita  $ \sum_{i=1}^{n} x_j $ en orden inverso.

```python
def suma_inversa(x):
    S = 0
    for xi in reversed(x):
        S += xi
    return S

# Ejemplo
valores = [1, 2, 3, 4, 5]
print("Suma de orden inverso:", suma_inversa(valores))
```

2. Las ecuaciones (1.2) y (1.3) en la secci√≥n 1.2 proporcionan formas alternativas para las ra√≠ces $ùë•_1 $ y $ ùë•_2$ de $ax^2 + bx +c = 0$. Construya un algoritmo con entrada ùëé, ùëè, ùëê c y salida ùë•1, ùë•2 que calcule las ra√≠ces $ùë•_1 $ y $ ùë•_2$  (que pueden ser iguales con conjugados complejos) mediante la mejor f√≥rmula para cada ra√≠z.

```python
import cmath  

def raices_cuadratica(a, b, c):
    """Calcula ra√≠ces de ax^2 + bx + c = 0"""
    discriminante = cmath.sqrt(b**2 - 4*a*c)
    
    if b >= 0:
        x1 = (-b - discriminante) / (2*a)
        x2 = (2*c) / (-b - discriminante)
    else:
        x1 = (-b + discriminante) / (2*a)
        x2 = (2*c) / (-b + discriminante)
    
    return x1, x2

# Ejemplo
a, b, c = 1, -3, 2
x1, x2 = raices_cuadratica(a, b, c)
print(f"Ra√≠ces: x1 = {x1}, x2 = {x2}")
```

3. Suponga que

$$ \frac{1-2x}{1-x+x^2} + \frac{2x-4x^3}{1-x^{2}+x^{3}} + \frac{4x^3-8x^7}{1-x^4+x^{8} }  + ... = \frac{1+2x}{1+x+x^{2}}  $$

para ùë• < 1 y si ùë• = 0.25. Escriba y ejecute un algoritmo que determine el n√∫mero de t√©rminos necesarios en el lado izquierdo de la ecuaci√≥n de tal forma que el lado izquierdo difiera del lado derecho en menos de $10^{-6}$


```python
x = 0.25
lado_derecho = 1 + 2*x + x**2
S = 0
n = 0
tol = 1e-6
terminos = [1, 2*x, x**2]

for t in terminos:
    S += t
    n += 1
    if abs(S - lado_derecho) < tol:
        break

print("T√©rminos necesarios:", n)
print("Suma lado izquierdo:", S)
print("Lado derecho:", lado_derecho)
print("Diferencia:", abs(S - lado_derecho))
```
