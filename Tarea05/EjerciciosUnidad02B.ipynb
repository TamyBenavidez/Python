## 2. Encuentre soluciones precisas dentro de $10^{-4}$ para los siguientes problemas.

```python
import math

# Funci√≥n para Newton
def newton(f, df, x0, tol=1e-4):
    x = x0
    while True:
        x_next = x - f(x)/df(x)
        if abs(x_next - x) < tol: break
        x = x_next
    return x

# Funci√≥n para Secante
def secante(f, p0, p1, tol=1e-4):
    while True:
        p2 = p1 - f(p1)*(p1-p0)/(f(p1)-f(p0))
        if abs(p2 - p1) < tol: break
        p0, p1 = p1, p2
    return p2
```

### a. $x^{3} - 2x^{2}-5 = 0$ , [1,4]

```python
f = lambda x: x**3 - 2*x**2 - 5
df = lambda x: 3*x**2 - 4*x
r2a = newton(f, df, 2)

print("Ra√≠z 2a ‚âà", r2a)
```

### b. $x^{3} - 3x^{2} - 1 = 0$, [-3,-2]

```python
f = lambda x: x**3 + 3*x**2 -1
df = lambda x: 3*x**2 + 6*x
r2b = newton(f, df, -2.5)

print("Ra√≠z 2b ‚âà", r2b)
```

###  c. $ x-cosx=0 $, [0,$\pi$/2]

## 3 Use los 2 m√©todos en esta secci√≥n para encontrar las soluciones dentro de $10^{-5}$ para los siguientes problemas.

### a. $3x-e^{x}=0$ para $1 \le  x \le 2$

```python
f = lambda x: 3*x - math.exp(x)
df = lambda x: 3 - math.exp(x)

# Newton
x = 1.5
while True:
    x_next = x - f(x)/df(x)
    if abs(x_next - x) < 1e-5: break
    x = x_next
print("Eje del m√©todo de Newton:", x)

# Secante
p0, p1 = 1,2
while True:
    p2 = p1 - f(p1)*(p1-p0)/(f(p1)-f(p0))
    if abs(p2 - p1) < 1e-5: break
    p0, p1 = p1, p2
print("Eje del m√©todo de la Secante:", p2)
```

### a. El m√©todo de la secante (use los extremos como las estimaciones iniciales)

### b. El m√©todo de Newton (use el punto medio como estimaci√≥n inicial)

```python
f = lambda x: 230*x**4 + 18*x**3 + 9*x**2 -221*x -9
df = lambda x: 920*x**3 + 54*x**2 + 18*x -221

# Secante [-1,0]
p0,p1 = -1,0
for _ in range(50):
    p2 = p1 - f(p1)*(p1-p0)/(f(p1)-f(p0))
    if abs(p2-p1)<1e-6: break
    p0,p1 = p1,p2
sec_neg = p2

# Newton [-1,0] punto medio
x = -0.5
while True:
    x_next = x - f(x)/df(x)
    if abs(x_next - x)<1e-6: break
    x = x_next
newt_neg = x

# Secante [0,1]
p0,p1 = 0,1
for _ in range(50):
    p2 = p1 - f(p1)*(p1-p0)/(f(p1)-f(p0))
    if abs(p2-p1)<1e-6: break
    p0,p1 = p1,p2
sec_pos = p2

# Newton [0,1]
x = 0.5
while True:
    x_next = x - f(x)/df(x)
    if abs(x_next - x)<1e-6: break
    x = x_next
newt_pos = x

print("Secante [-1,0]:",sec_neg,"Newton [-1,0]:",newt_neg)
print("Secante [0,1]:",sec_pos,"Newton [0,1]:",newt_pos)
```

### 5. La funcion $f(x)=tan(\pi x)-6 $ tiene cero en $1 / \pi arcontangente 6 \approx 0.447431543$. Sea $p_0 = 0$ y $p_1 = 0.48$ y use 10 iteraciones en cada uno de los siguientes m√©todos para aproximar esta ra√≠z. ¬øCu√°l m√©todo es m√°s eficaz y por qu√©?

### a. m√©todo de bisecci√≥n

```python
f = lambda x: math.tan(math.pi*x)-6
df = lambda x: math.pi/(math.cos(math.pi*x)**2)

a,b = 0,0.48
for _ in range(10):
    c = (a+b)/2
    if f(a)*f(c)<=0: b=c
    else: a=c
bisec = (a+b)/2

print("Bisecci√≥n ‚âà", bisec)
```

### b. m√©todo de Newton

```python
x=0
for _ in range(10):
    x = x - f(x)/df(x)

print("Newton ‚âà", x)
```

### c. m√©todo de la secante

```python
p0,p1 = 0,0.48
for _ in range(10):
    p2 = p1 - f(p1)*(p1-p0)/(f(p1)-f(p0))
    p0,p1 = p1,p2

print("Secante ‚âà", p2)
```

## 6. La funci√≥n descrita por $f(x) = ln(x^{2}+1) -e^{0.4x} cosùúãùë•$ tiene un n√∫mero infinito de ceros.

```python
f = lambda x: math.log(x**2+1)-math.exp(0.4*x)*math.cos(math.pi*x)

def bisection(f,a,b,tol=1e-6):
    while b-a>tol:
        c=(a+b)/2
        if f(a)*f(c)<=0: b=c
        else: a=c
    return (a+b)/2
```

### a. Determine, dentro de $10^{-6}$, el √∫nico cero negativo.

```python
neg = bisection(f,-1,0)
print("Cero negativo ‚âà", neg)
```

### b. Determine, dentro de $10^{‚àí6}$, los cuatro ceros positivos m√°s peque√±os

```python
pos = [bisection(f,i,i+1) for i in range(5)]
print("Ceros positivos ‚âà", pos)
```

## La funci√≥n $ùëì(x) = x^{1/3}, \quad \text{ra√≠z en } x=0$.Usando el punto de inicio de ùë•=1 y ùëù0=5, ùëù1=0.5 para el m√©todo de secante, compare los resultados de los m√©todos de la secante y de Newt
