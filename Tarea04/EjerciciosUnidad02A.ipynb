1. Use el mÃ©todo de bisecciÃ³n para encontrar soluciones precisas dentro de  $ 10^{-2}$ para $ x^{3} -7x^{2} + 14x-6=0$  en cada intervalo.

    a. [0, 1]
    
    b. [1, 3.2]
    
    c. [3.2, 4]

```python
def biseccion(f, a, b, tol=1e-2, max_iter=100):
    
    if f(a)*f(b) > 0:
        return None, 0  # No hay cambio de signo
    
    for i in range(max_iter):
        c = (a + b)/2
        if f(c) == 0 or (b - a)/2 < tol:
            return c, i+1
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
    return (a+b)/2, max_iter

# FunciÃ³n
f = lambda x: x**3 - 7*x**2 + 14*x - 6

# Intervalos
intervalos = {
    "a": (0, 1),
    "b": (1, 3.2),
    "c": (3.2, 4)
}

# Se aplica el mÃ©todo de BisecciÃ³n a cada intervalo
for key, (a, b) in intervalos.items():
    raiz, iteraciones = biseccion(f, a, b)
    print(f"RaÃ­z {key}: {raiz:.4f} encontrada en {iteraciones} iteraciones")
```

$V = L \left[ \frac{1}{2} \pi r^{2} - r^{2} \arcsin\left(\frac{h}{r}\right) - h \sqrt{r^2 - h^2} \right]$

![](EjercicioA1.png){width=7cm height=5cm fig-align="center"}

Suponga que ð¿ = 10 ð‘ð‘š, ð‘Ÿ = 1 ð‘ð‘š y ð‘‰ = 12.4 $cm^{3} $. Encuentre la profundidad del agua en el abrevadero dentro de 0.01 ð‘ð‘š.


```python
import numpy as np

# Datos
L = 10  # cm
r = 1   # cm
V_obj = 12.4  # cm^3

# FunciÃ³n f(h) = V(h) - V_obj
def f(h):
    return L*(0.5*np.pi*r**2 - r**2*np.arcsin(h/r) - h*np.sqrt(r**2 - h**2)) - V_obj

# MÃ©todo de BisecciÃ³n
def biseccion(f, a, b, tol=0.01, max_iter=100):
    if f(a)*f(b) > 0:
        return None, 0  # No hay cambio de signo
    for i in range(max_iter):
        c = (a + b)/2
        if f(c) == 0 or (b - a)/2 < tol:
            return c, i+1
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
    return (a+b)/2, max_iter

# Intervalo posible [0, r]
raiz, iteraciones = biseccion(f, 0, r)
print(f"Profundidad del agua: h = {raiz:.2f} cm encontrada en {iteraciones} iteraciones")
```

2.  Un objeto que cae verticalmente a travÃ©s del aire estÃ¡ sujeto a una resistencia viscosa, asÃ­ como a la fuerza de gravedad. Suponga que un objeto con masa ð‘š cae desde una altura  $S_0$  y que la altura del objeto despuÃ©s de ð‘¡ segundos es

$S(t) = S_0 - \frac{mg}{k} t + \frac{m^2 g}{k^2} \left( 1 - e^{- \frac{k}{m} t} \right)$

donde g = 9.81 $m/s^{2}$ y k representa el coeficiente de la resistencia del aire en $Ns/m$. Suponga $S_0$ = 300m, m = 0.25 kg y k = 0.1 Ns/m. Encuentre dentro de 0.01 ð‘ ð‘’ð‘”ð‘¢ð‘›ð‘‘ð‘œð‘ , el tiempo que tarda un cuarto de kg en golpear el piso.

```python
import numpy as np

# Datos
s0 = 300   # m
m = 0.25   # kg
k = 0.1    # NÂ·s/m
g = 9.81   # m/s^2

# FunciÃ³n f(t) = s(t)
def f(t):
    return s0 - (m*g/k)*t + (m**2 * g / k**2)*(1 - np.exp(-k*t/m))

# MÃ©todo de bisecciÃ³n
def biseccion(f, a, b, tol=0.01, max_iter=100):
    if f(a)*f(b) > 0:
        return None, 0  # No hay cambio de signo
    for i in range(max_iter):
        c = (a + b)/2
        if f(c) == 0 or (b - a)/2 < tol:
            return c, i+1
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
    return (a+b)/2, max_iter

# Intervalo aproximado [0, 100] s
raiz, iteraciones = biseccion(f, 0, 100)
print(f"Tiempo para caer al piso: t = {raiz:.2f} s en {iteraciones} iteraciones")
```

1.	Use el teorema 2.1 para encontrar una cota para el nÃºmero de iteraciones necesarias para lograr una aproximaciÃ³n con precisiÃ³n de $10^{-4}$ para la soluciÃ³n de $x^{3}-x-1=0$ que se encuentra dentro del intervalo [1, 2]. Encuentre una aproximaciÃ³n para la raÃ­z con este grado de precisiÃ³n

```python
import numpy as np

# Datos
a = 1
b = 2
tol = 1e-4

N_max = np.ceil((np.log(b - a) - np.log(tol)) / np.log(2))
print(f"Cota mÃ¡xima de iteraciones: N â‰¤ {N_max:.0f}")

# MÃ©todo de bisecciÃ³n
f = lambda x: x**3 - x - 1

def biseccion(f, a, b, tol=1e-4, max_iter=100):
    if f(a)*f(b) > 0:
        return None, 0
    for i in range(max_iter):
        c = (a + b)/2
        if f(c) == 0 or (b - a)/2 < tol:
            return c, i+1
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
    return (a+b)/2, max_iter

raiz, iteraciones = biseccion(f, a, b, tol=tol, max_iter=int(N_max))
print(f"RaÃ­z aproximada: x â‰ˆ {raiz:.4f} encontrada en {iteraciones} iteraciones")
```
